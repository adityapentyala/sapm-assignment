```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
df=pd.read_csv('data/returns.csv',index_col=0,parse_dates=True)
```

```python
aapl = yf.download("AAPL", start="2024-10-10", end="2025-10-10", progress=False, group_by='ticker')
print(aapl.tail())
```

```python
aapl.columns = ['_'.join(col).strip() for col in aapl.columns.values]
aapl.head()
```

```python
aapl['Return'] = aapl['AAPL_Close'].pct_change()
aapl = aapl[['Return']].rename(columns={'Return': 'AAPL'})
```

```python
df.index = df.index.date
```

```python
combined_df = df.join(aapl, how='inner')
```

```python
combined_df.index.name = "date"
```

```python
combined_df.to_csv("data/returns_with_international.csv")
```
